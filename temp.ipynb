```python
# Resize image

from PIL import Image
import matplotlib.pyplot as plt

bg = Image.open("images/bg_0.jpg")
plt.imshow(bg)

width, height = bg.size

newsize = (int(width*0.5), int(height*0.5))
im1 = bg.resize(newsize)

plt.imshow(im1)
```

```python
# Cropped image of above dimension
# (It will not change original image)
im2 = im1.crop((0, 0, 1274, 1274))

plt.imshow(im2)

im2.save("images/bg_0_cropped.jpg")
```
